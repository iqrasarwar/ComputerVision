```python
                                                    #BSEF19M012 - TAKEHOME QUIZ 02
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Reading in Grey scale mode
img = cv2.imread('/content/drive/MyDrive/ComputerVisionCoLabFiles/cat.jfif', 0) 
plt.imshow(img, cmap='gray')
```

```python
                                                      #Calculating Threshold By Averaging Method
THRESH = img.mean()
while True:
    foreground_img = (img > THRESH) * img
    background_img = (img <= THRESH) * img
    mean1 = foreground_img.mean()   
    mean2 = background_img.mean() 
    new_THRESH = (mean1 + mean2) / 2    
    diff = new_THRESH - THRESH          
    
    if diff > 1e-4:            
        THRESH = new_THRESH
    else:                      
        break
```

```python
# Creating image size empty images to hold result for thresholding
tt_img= np.zeros((r,c),np.uint8)  #Truncate Thresholding
zi_img= np.zeros((r,c),np.uint8)  #Zero Inverted Thresholding
ib_img= np.zeros((r,c),np.uint8)  #Inverted Binary Thresholding
# Getting rows and cols
r,c = img.shape
```

```python
                                                              # Truncate Thresholding
for i in range(r):
    for j in range(c):
        if(img[i,j] > THRESH):
            tt_img[i,j] = THRESH
        else:
            tt_img[i,j]= img[i,j]
cv2.imwrite("Truncate_Thresholding_Image.jpg", tt_img)
plt.imshow(tt_img, cmap='gray')
```

```python
                                                              # Zero Inverted Thresholding
for i in range(r):
    for j in range(c):
        if(img[i,j] > THRESH):
            zi_img[i,j] = 0
        else:
            zi_img[i,j]= img[i,j]
cv2.imwrite("Zero_Inverted_Thresholding_Image.jpg", zi_img)
plt.imshow(zi_img, cmap='gray')
```

```python
                                                              # Inverted Binary Thresholding
for i in range(r):
    for j in range(c):
        if(img[i,j] > THRESH):
            ib_img[i,j] = 0
        else:
            ib_img[i,j]= 255
cv2.imwrite("Inverted_Binary_Thresholding_Image.jpg", ib_img)
plt.imshow(ib_img, cmap='gray')
```
