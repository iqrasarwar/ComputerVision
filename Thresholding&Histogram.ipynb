```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
OldImg = cv.imread("/content/drive/MyDrive/ComputerVisionCoLabFiles/cat.jfif",0);
```

```python
# FOR THRESHOLDING READ IMAGE IN GRAY SCALE
thresh = 125
rows,cols = OldImg.shape
i = 0
j = 0
newimg = np.zeros((rows,cols),dtype = np.uint8)
for i in range ( rows):
  for j in range ( cols ):
    if( OldImg [ i , j ] > thresh ):
      newimg [ i , j ] = 1
    else:
      newimg[ i , j ] = 0
plt.imshow(newimg , cmap = "gray")
plt.show()
```

```python
plt.hist(OldImg , bins =10)
plt.show()
```

```python
OldImg = cv.equalizeHist(OldImg)
plt.hist(OldImg , bins =10)
plt.show()
```
