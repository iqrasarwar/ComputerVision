```python
arr = np.array([1,2,3,3])
print(type(arr))
print(arr.ndim)
print(arr.size)
print(arr.shape)
print(arr.dtype)
```

```python
#shape ?
arr = np.array([[1,2,3],
                [4,5,6],
                [3,6,7]])
print(type(arr))
print(arr.ndim)
print(arr.size)
print(arr.shape)
print(arr.dtype)
print(arr)
```

```python
arr = np.zeros((2,2))
print(type(arr))
print(arr.ndim)
print(arr.size)
print(arr.shape)
print(arr.dtype)
print(arr)
```

```python
arr = np.ones((3,3))
print(type(arr))
print(arr.ndim)
print(arr.size)
print(arr.shape)
print(arr.dtype)
print(arr)
```

```python
import cv2 as cv
import matplotlib.pylab as plt
```

```python
arr = plt.imread("cat.jfif")
plt.imshow(arr)
type(arr)
```

```python
print(type(arr))
print(arr.ndim)
print(arr.size)
print(arr.shape)
print(arr.dtype)
print(arr.min())
print((arr.max()))
print(arr.mean()) #avg
```

```python
arr = plt.imread("cat.jfif")
print(arr[13,10])
arr = arr[13,10]
print(arr)
print(arr.ndim)
print(arr.size)
print(arr.shape)
print(arr.dtype)
print(arr.min())
print((arr.max()))
print(arr.mean())
```
