```python
import matplotlib.pylab as plt
import numpy as np
import cv2 as cv
import math
```

```python
cat = plt.imread("/content/drive/MyDrive/ComputerVisionCoLabFiles/cat.jfif")
```

```python
plt.imshow(cat)
```

```python
rows = cat.shape[0]
columns = cat.shape[1]
grayImg = np.zeros((rows,columns))
for i in range(rows):
  for j in range(columns):
    red = cat[i,j,0]
    green = cat[i,j,1]
    blue = cat[i,j,2]
    gray = (((0.21*red)+(0.07*blue)+(0.72*green))/3)
    grayImg[i][j] = gray
grayCat = cv.imwrite('grayCat.jpeg', grayImg)
grayCat = plt.imread('grayCat.jpeg')
plt.imshow(grayCat,cmap='gray')
```
